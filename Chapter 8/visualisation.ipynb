```python
import pandas as pd

file = "../chapter_3_assets/8_matplotlib_seaborn/linechart.xlsx"

df_sales = pd.read_excel(file, sheet_name='product sales')
df_sales.head()
```

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(15, 6))
plt.title("Product Sales Per Quarter")
plt.plot('Quarter', 'Dishwasher', 'r+', data=df_sales, linestyle='-', label='Dishwasher')
plt.plot('Quarter', 'Fridge', 'go', data=df_sales, linestyle='--', label='Fridge')
plt.plot('Quarter', 'Washing Machine', 'y*', data=df_sales, linestyle='-', label='Washing Machine')
plt.xlabel("Quarter")
plt.ylabel("Sales in million ($)")
plt.legend()
plt.show()
```

```python
sum_all = df_sales[['Fridge', 'Dishwasher', 'Washing Machine']].sum()
sum_all.keys()
```

```python
fig, ax = plt.subplots(figsize=(18, 8))
ax.pie(sum_all, labels=sum_all.keys(), autopct='%1.1f%%', explode=(0, 0.1, 0), shadow=True, startangle=50)
ax.set_title("Total Revenue Per Item")
ax.legend(loc='lower right')

plt.show()
```

```python
df_sales.plot(kind='bar', x='Quarter', figsize=(18, 8))
plt.xticks(rotation=45)
plt.ylabel('Sales in millions ($)')
plt.title('Sales Per Quarter from 2022')
plt.show()
```
